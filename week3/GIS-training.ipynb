```python
import geopandas as gp
import seaborn as sns
sns.set()
world = gp.read_file("world/world_m.shp")
print(world.head())
```

```python
world.plot(column='name')
```

```python
cities = gp.read_file("cities/cities.shp")
print(cities.head())
```

```python
cities.plot(markersize=5)
```

```python
cities = cities.to_crs({'init': 'epsg:3395'})
cities.plot(markersize=5)
```

```python
base = world.plot(color='white', edgecolor='black')
cities.plot(ax=base, marker='o', color='red', markersize=2)
```
